```python
# Imports
from pkg_resources import resource_filename
import os
import numpy as np
from astropy.io import fits
from astropy.wcs import WCS
from astropy.coordinates import SkyCoord
from astropy.wcs.utils import pixel_to_skycoord, skycoord_to_pixel
from astropy.table import Table

# Photometry
from astropy.stats import sigma_clipped_stats
from photutils import DAOStarFinder

from photutils.psf import IterativelySubtractedPSFPhotometry
from photutils.psf import BasicPSFPhotometry
from photutils.psf import IntegratedGaussianPRF, DAOGroup
from photutils.background import MMMBackground, MADStdBackgroundRMS
from astropy.modeling.fitting import LevMarLSQFitter
from astropy.stats import gaussian_sigma_to_fwhm

# ASTR 257
from astr257 import img_utils

# Personal utils
from utils import *

# Matplotlib
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams["font.family"] = "serif"
rcParams["font.serif"] = "Times New Roman"
%matplotlib inline
#%matplotlib notebook
rcParams['text.usetex'] = True
rcParams['text.latex.preamble'] = [r'\usepackage{amsmath} \usepackage{bm} \usepackage{physics}']
%config InlineBackend.figure_format = 'retina' # For high quality figures
%load_ext autoreload
%autoreload 2
```

```python
data_folder = "data-2019-09-23-nickel-joey-anne"
OVERWRITE = True
```

# Load the images

```python
ngc6819_v = fits.getdata(os.path.join(data_folder, 'masters/ngc6819_v_master_full_cal_nobadcol.fits'))
ngc6819_v_header = fits.getheader(os.path.join(data_folder, 'masters/ngc6819_v_master_full_cal_nobadcol.fits'))

ngc6819_b = fits.getdata(os.path.join(data_folder, 'masters/ngc6819_b_master_full_cal_nobadcol.fits'))
ngc6819_b_header = fits.getheader(os.path.join(data_folder, 'masters/ngc6819_b_master_full_cal_nobadcol.fits'))

landolt_v = fits.getdata(os.path.join(data_folder, 'masters/landolt_v_master_full_cal_nobadcol.fits'))
landolt_v_header = fits.getheader(os.path.join(data_folder, 'masters/landolt_v_master_full_cal_nobadcol.fits'))

landolt_b = fits.getdata(os.path.join(data_folder, 'masters/landolt_b_master_full_cal_nobadcol.fits'))
landolt_b_header = fits.getheader(os.path.join(data_folder, 'masters/landolt_b_master_full_cal_nobadcol.fits'))
```

## Perform iterative PSF Photometry on NGC6819 V image

```python
mean, median, std = sigma_clipped_stats(ngc6819_v, sigma=3.0) # Background statistics
daofinder = DAOStarFinder(fwhm=4.0, threshold=3.*std, 
                          sharplo=0.0, sharphi=0.5, 
                          roundlo=-0.2, roundhi=0.2, 
                          peakmax=3e4, exclude_border=True)
sigma_psf = 2.0
daogroup = DAOGroup(2.0*sigma_psf*gaussian_sigma_to_fwhm)
mmm_bkg = MMMBackground()
fitter = LevMarLSQFitter()
psf_model = IntegratedGaussianPRF(sigma=sigma_psf)

# Changing the number of iterations can affect B-V color significantly
photometry = IterativelySubtractedPSFPhotometry(finder=daofinder, 
                                                group_maker=daogroup, 
                                                bkg_estimator=mmm_bkg, 
                                                psf_model=psf_model, 
                                                fitter=LevMarLSQFitter(), 
                                                niters=1,
                                                fitshape=(11,11))
result_tab = photometry(image=ngc6819_v)
residual_image = photometry.get_residual_image()
```

```python
wcs = WCS(ngc6819_v_header)
plt.figure(figsize=(10, 10))
plt.subplot(projection=wcs)
plt.imshow(residual_image, vmin=0, vmax=3000)
plt.grid(color='white', ls='solid')
plt.show()
```

## Convert pixels in V image to SkyCoords, then convert to pixels in B image

```python
ngc6819_sources_v_skycoords = pixel_to_skycoord(result_tab['x_0'], result_tab['y_0'], wcs)
ngc6819_sources_b_pixels    = skycoord_to_pixel(ngc6819_sources_v_skycoords, WCS(ngc6819_b_header))
ngc6819_sources_b_tab = Table((ngc6819_sources_b_pixels[0], ngc6819_sources_b_pixels[1]), names=['x_0', 'y_0'])
```

## Now run basic psf on the NGC6819 B image

```python
sigma_psf = 2.0
daogroup = DAOGroup(2.0*sigma_psf*gaussian_sigma_to_fwhm)
mmm_bkg = MMMBackground()
psf_model = IntegratedGaussianPRF(sigma=sigma_psf)

photometry_b = BasicPSFPhotometry(group_maker=daogroup, 
                                  bkg_estimator=mmm_bkg, 
                                  psf_model=psf_model,
                                  finder=None,
                                  fitshape=((11,11)))
result_tab_b = photometry_b(ngc6819_b, init_guesses=ngc6819_sources_b_tab)
residual_image_b = photometry_b.get_residual_image()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(projection=WCS(ngc6819_b_header))
plt.imshow(residual_image_b, vmin=0, vmax=3000)
plt.grid(color='white', ls='solid')
plt.show()
```

## Data cleaning 
#### Some 'flux_fit' values in the results table are negative, so throw out those rows. 

```python
result_tab = result_tab[result_tab['flux_fit'] > 0]
result_tab_b = result_tab_b[result_tab_b['flux_fit'] > 0]
```

### Some rows have a flux uncertainty of 0.0. Looking at their flux counts, they're saturated (in V)

```python
# Easier to do this in Pandas
result_df = result_tab.to_pandas()
result_df_b = result_tab_b.to_pandas()
```

```python
# Drop the saturated rows
saturated_rows = result_df.index[result_df['flux_unc'] == 0].tolist()
result_df = result_df.drop(result_df.index[saturated_rows])
result_df_b = result_df_b.drop(result_df_b.index[saturated_rows])
```

### Check to see if the tables are the same length
If not, they'll be fixed in the source_matching notebook

```python
try:
    assert len(result_df) == len(result_df_b), " "
    print("The tables are the same length.")
except:
    print("len(V) = {}, len(B) = {}.".format(len(result_df), len(result_df_b)))
```

```python
# Convert back to astropy Tables
result_tab = Table.from_pandas(result_df)
result_tab_b = Table.from_pandas(result_df_b)
```

## Save the data tables with the PSF photometry results for the science images

```python
print('OVERWRITE = {}'.format(OVERWRITE))
```

```python
try:
    result_tab.write(os.path.join(data_folder, 'photometry/ngc6819_v_iter_1.dat'), format='ascii', 
                     overwrite=OVERWRITE)
    result_tab_b.write(os.path.join(data_folder, 'photometry/ngc6819_b_iter_1.dat'), format='ascii', 
                       overwrite=OVERWRITE)
except:
    print('Warning: Did not overwrite file.')
```

# Photometry on the Landolt fields

```python
landolt_mean, landolt_median, landolt_std = sigma_clipped_stats(landolt_v, sigma=3.0) # Background statistics
landolt_daofinder = DAOStarFinder(fwhm=3.0, 
                          threshold=20.*landolt_std, 
                          sharplo=0.0, sharphi=.75, 
                          roundlo=-0.4, roundhi=0.4, 
                          peakmax=5e4, exclude_border=True)
sigma_psf = 2.0
daogroup = DAOGroup(2.0*sigma_psf*gaussian_sigma_to_fwhm)
mmm_bkg = MMMBackground()
fitter = LevMarLSQFitter()
psf_model = IntegratedGaussianPRF(sigma=sigma_psf)
photometry = IterativelySubtractedPSFPhotometry(finder=landolt_daofinder, 
                                                group_maker=daogroup, 
                                                bkg_estimator=mmm_bkg, 
                                                psf_model=psf_model, 
                                                fitter=LevMarLSQFitter(), 
                                                niters=3, 
                                                fitshape=(11,11))
result_tab_landolt = photometry(image=landolt_v)
residual_image_landolt = photometry.get_residual_image()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(projection=WCS(landolt_v_header))
plt.imshow(residual_image_landolt, vmin=0, vmax=3000)
plt.grid(color='white', ls='solid')
plt.show()
```

## Convert to skycoordinates and back to pixels

```python
landolt_sources_v_skycoords = pixel_to_skycoord(result_tab_landolt['x_0'], 
                                                result_tab_landolt['y_0'], 
                                                WCS(landolt_v_header))
landolt_sources_b_pixels    = skycoord_to_pixel(landolt_sources_v_skycoords, WCS(landolt_b_header))
landolt_sources_b_tab = Table((landolt_sources_b_pixels[0], landolt_sources_b_pixels[1]), names=['x_0', 'y_0'])
```

```python
sigma_psf = 2.0
daogroup = DAOGroup(2.0*sigma_psf*gaussian_sigma_to_fwhm)
mmm_bkg = MMMBackground()
psf_model = IntegratedGaussianPRF(sigma=sigma_psf)

photometry_landolt_b = BasicPSFPhotometry(group_maker=daogroup, 
                                  bkg_estimator=mmm_bkg, 
                                  psf_model=psf_model,
                                  finder=None,
                                  fitshape=((11,11)))
result_tab_landolt_b = photometry_landolt_b(landolt_b, init_guesses=landolt_sources_b_tab)
residual_image_landolt_b = photometry_landolt_b.get_residual_image()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(projection=WCS(landolt_b_header))
plt.imshow(residual_image_landolt_b, vmin=0, vmax=3000)
plt.grid(color='white', ls='solid')
plt.show()
```
