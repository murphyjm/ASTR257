```python
import os
import numpy as np
from astropy.io import fits
from linetools.spectra.xspectrum1d import XSpectrum1D

# Matplotlib
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams["font.family"] = "serif"
rcParams["font.serif"] = "Times New Roman"
%matplotlib inline
rcParams['text.usetex'] = True
rcParams['text.latex.preamble'] = [r'\usepackage{amsmath} \usepackage{bm} \usepackage{physics}']
%config InlineBackend.figure_format = 'retina' # For high quality figures
```

```python
data_folder = "../data-2019-09-25-shane-joey-alex-arcelia"
```

```python
OVERWRITE = True
```

## Load the arc spectrum

```python
# Load the data and headers
arc_fname = os.path.join(data_folder, 'masters/arcs_combined.fits')
arc_spec, arc_header = fits.getdata(arc_fname, header=True)
```

```python
# Median across rows (median flux at each wavelength)
row_num  = np.arange(arc_spec.shape[0])
arc_fl  = np.median(arc_spec, axis=1)
```

```python
# Take a quick look
fig, ax = plt.subplots(figsize=(9,6))
ax.plot(row_num, arc_fl)
plt.show()
```

## Detrend the arc spectrum

```python
z = np.polyfit(row_num, arc_fl, 2)
p = np.poly1d(z)
```

```python
# Check the detrend
fig, axes = plt.subplots(nrows=2, figsize=(9,6), sharex=True)
ax0, ax1 = axes.flatten()

x_grid = np.linspace(0.95 * np.min(row_num), 1.05 * np.max(row_num), 2000)

ax0.plot(row_num, arc_fl)
ax0.plot(x_grid, p(x_grid))

ax1.plot(row_num, arc_fl - p(row_num))
ax1.plot(x_grid, 100 * np.ones(len(x_grid)), '--', color='gray', alpha=0.5)

plt.show()
```

```python
# Store the detrended arc spectrum
arc_fl_detrend = arc_fl - p(row_num)
```

## Find arc lines by pixel

```python
from scipy.signal import find_peaks
```

```python
peak_inds = find_peaks(arc_fl_detrend, height=100)[0]
print(peak_inds)
```

## Match the peaks at the pixels above to the arc lamp spectrum
Here: https://mthamilton.ucolick.org/techdocs/instruments/kast/images/Kastred1200NeAr_maxtilt.jpg

```python
# True wls corresponding to the peaks of the indices above^^
true_wls = np.array([6143.062, 6266.495, 6334.428, 6382.991, 6402.246, 6506.528, 7032.413, 7245.167])

x = np.copy(peak_inds)
y = np.copy(true_wls)
```

## Fit a polynomial to the pixel vs wavelength data

```python
deg = 3
z1  = np.polyfit(x, y, deg)
p1  = np.poly1d(z1)
```

```python
fig, axes = plt.subplots(nrows=2, figsize=(9,5), sharex=True)
ax0, ax1 = axes.flatten()

x_grid = np.linspace(0.95 * np.min(row_num), 1.05 * np.max(row_num), 2000)

ax0.plot(x_grid, p1(x_grid), label='Fit')
ax0.plot(x, y, '.r', label='Data')

ax1.plot(x_grid, np.zeros(x_grid.shape[0]), '--', color='gray', alpha=0.5)
ax1.plot(x, y - p1(x), '.', label='Residuals')

ax0.legend()
ax1.legend()
ax0.set_ylabel(r'$\lambda$ [Angstroms]')
ax1.set_ylabel(r'$\lambda$ [Angstroms]')
ax1.set_xlabel('Row Number')

plt.show()
```

# Wavelength calibrate the UGC 11876 spectrum

## Load the data

```python
ugc11876, header_ugc11876 = fits.getdata(os.path.join(data_folder, 'masters/ugc11876_master.fits'), header=True)
```

```python
# Median across rows
ugc11876_fl = np.median(ugc11876, axis=1)
```

```python
# Convert row numbers to wavelengths
wl = p1(row_num)
```

```python
# Take a quick look at the solution
fig, ax = plt.subplots(figsize=(9,6))

ax.plot(wl, ugc11876_fl)
ax.set_xlabel(r'$\lambda$')
ax.set_ylabel('Counts')
ax.set_ylim([900, 920])

plt.show()
```

```python
# Take a longer look with linetools
ugc_spec = XSpectrum1D.from_tuple((wl, ugc11876_fl))
ugc_spec.plot(xspec=True)
```
