```python
import os
import numpy as np
from astropy.io import fits
from linetools.spectra.xspectrum1d import XSpectrum1D
from scipy.signal import find_peaks

# Matplotlib
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams["font.family"] = "serif"
rcParams["font.serif"] = "Times New Roman"
%matplotlib inline
rcParams['text.usetex'] = True
rcParams['text.latex.preamble'] = [r'\usepackage{amsmath} \usepackage{bm} \usepackage{physics}']
%config InlineBackend.figure_format = 'retina' # For high quality figures
```

# Load the wavelength solution

```python
z  = np.load('../wl_func_coeffs.npy')
p  = np.poly1d(z)
wl = np.load('../wl.npy')
```

## Load the UGC11876 data

```python
ugc11876, header_ugc11876 = fits.getdata(os.path.join(data_folder, 'masters/ugc11876_master.fits'), header=True)
```

```python
# Take a quick look at the spectra, taking the mean across pixel

fig, ax = plt.subplots(figsize=(9,6))
ax.plot(wl, np.mean(ugc11876, axis=1))
ax.set_xlabel(r'$\lambda$ [Angstroms]', fontsize=14)
ax.set_ylabel('Counts', fontsize=14)
plt.show()
```

## Now locate the centroids of $H_\alpha$ at each pixel (radius)

```python
# Plot every spectrum of each pixel, zooming in on H_alpha region
fig, ax = plt.subplots(figsize=(9,6))
ax.plot(wl, ugc11876)
ax.set_xlabel(r'$\lambda$ [Angstroms]', fontsize=14)
ax.set_ylabel('Counts', fontsize=14)
ax.set_xlim([6645, 6657.5])  # Angstroms
ax.set_ylim([780, 900])      # Counts
plt.show()
```

```python
# Map this zoomed in region from wavelength number to column index in the ugc11876 image
blue_ind = np.argwhere(np.isclose(wl, 6647 * np.ones(len(wl)), rtol=1e-4))[0][0]
red_ind  = np.argwhere(np.isclose(wl, 6656 * np.ones(len(wl)), rtol=1e-4))[0][0]
```

### Use find_peaks() to find the wavelength at the ~center of each H_alpha peak for each pixel

The wl_peak_at_pixel array is of shape (ugc11876.shape[0], 2). The first column contains the peak centroid, the second column contains the pixel number wrt to the original ugc11876 science image.

```python
wl_peak_at_pixel = np.zeros((ugc11876.shape[0], 2))

for pixel in range(ugc11876.shape[1]):
    peak_inds_pixel = find_peaks(ugc11876[blue_ind:red_ind+1, pixel], height=810)[0]
    if peak_inds_pixel.shape[0] == 0:
        continue
    elif peak_inds_pixel.shape[0] > 1:
        peak_inds_pixel = np.mean(peak_inds_pixel)
        wl_peak = p(peak_inds_pixel + blue_ind)
        wl_peak_at_pixel[pixel] = np.array([wl_peak, pixel])
    elif peak_inds_pixel.shape[0] == 1:
        wl_peak = p(peak_inds_pixel + blue_ind) # To map back to original ugc11876 array indices
        wl_peak_at_pixel[pixel] = np.array([wl_peak[0], pixel])
```

### Plot the spectra for each pixel next to a histogram of where each pixel's spectra's H_alpha peak was found

```python
fig, axes = plt.subplots(nrows=2, figsize=(9,7), sharex=True)
ax0, ax1 = axes.flatten()
ax0.plot(wl, ugc11876)
ax0.set_xlim([6645, 6657.5])  # Angstroms
ax0.set_ylim([780, 900])      # Counts
ax0.set_ylabel('Counts', fontsize=14)

ax1.hist(wl_peak_at_pixel[wl_peak_at_pixel[:, 0] > 0.][:, 0], bins=20)
ax1.set_xlabel(r'$\lambda$ [Angstroms]', fontsize=14)
ax1.set_ylabel('N', fontsize=14)

plt.show()
```
